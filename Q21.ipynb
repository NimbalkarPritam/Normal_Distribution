```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CR=pd.read_csv("Downloads\Cars.csv")
```

```python
CR.MPG.mean()
```

```python
CR.MPG.median()
```

```python
CR.MPG.mode()
```

```python
CR.MPG.std()
```

```python
%matplotlib inline
plt.plot(CR.MPG);
plt.xlabel('Distance');
plt.ylabel('Miles');
```

```python
sns.distplot(CR.MPG, label = 'INFO');
plt.xlabel('Distance');
plt.ylabel('Miles');
plt.legend();
```
